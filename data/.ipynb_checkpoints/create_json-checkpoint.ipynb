# Create country JSON

```python
import pandas as pd
import numpy as np
import os

# Get tables and define countries to exclude

base_table = pd.read_html('https://www.geonames.org/countries/', keep_default_na=False, encoding='utf-8')[1]
country_names_german = pd.read_html('https://www.laenderdaten.de/laender.aspx', keep_default_na=False, encoding='utf-8')[0][['Land', 'ISO-2']]
capitals_german = pd.read_html('https://www.laenderdaten.de/staat/hauptstaedte.aspx', keep_default_na=False, encoding='utf-8')[0][['Land', 'Hauptstadt']]
country_levels = pd.read_csv('country_levels.csv', sep=';')
not_countries = pd.read_csv('not_countries.csv', sep=';')['countryCode'].tolist()
svg_not_existing = ['fm', 'mh', 'tv', 'xk']
countries_to_drop = not_countries + svg_not_existing

# Rename columns and clean up data

countries_table = base_table[['ISO-3166alpha2', 'Country', 'Capital', 'Continent']]
countries_table.rename(columns = {'ISO-3166alpha2':'countryCode', 'Country':'nameEn', 'Capital':'capitalEn', 'Continent':'continent'}, inplace = True)
capitals_german.rename(columns = {'Land':'nameGer', 'Hauptstadt':'capitalGer'}, inplace = True)
country_names_german.rename(columns = {'Land':'nameGer', 'ISO-2':'countryCode'}, inplace=True)
country_names_german = country_names_german[country_names_german['nameGer'] != 'Clipperton']

for index, row in country_names_german.iterrows():
    if not pd.isnull(row['countryCode']):
        country_names_german.at[index, 'countryCode'] = row['countryCode'].lower()
countries_table['filename'] = ''

countries_table['continent'] = countries_table['continent'].replace(
    to_replace=['EU', 'AS', 'OC', 'AF', 'NA', 'SA'],
    value=['Europe', 'Asia', 'Oceania', 'Africa', 'North America', 'South America'])

for index, row in countries_table.iterrows():
    countries_table.at[index, 'countryCode'] = row['countryCode'].lower()
    row['filename'] = row['countryCode'] + '.svg'
    if pd.isnull(row['continent']):
        countries_table.at[index, 'continent'] = 'North America'

countries_table = countries_table[~countries_table.countryCode.isin(countries_to_drop)]

countries_table = pd.merge(countries_table, country_names_german, on = "countryCode", how = "outer")
countries_table = pd.merge(countries_table, capitals_german, on = "nameGer", how = "outer")
countries_table = pd.merge(countries_table, country_levels, on = "countryCode", how = "outer")

countries_table = countries_table[['countryCode', 'nameEn', 'nameGer', 'capitalEn', 'capitalGer', 'continent', 'level', 'filename']]
countries_table = countries_table.dropna()
countries_table = countries_table.replace(to_replace = '\s\([^)]*\)', value = '', regex = True)
countries_table = countries_table.replace(to_replace = 'Côte d\'Ivoire', value = 'Elfenbeinküste', regex = True)
countries_table = countries_table.replace(to_replace = 'St. George´s', value = 'St. George\'s', regex = True)
countries_table = countries_table.replace(to_replace = 'keine offizielle Hauptstadt, inoffiziell Yaren', value = 'Yaren', regex = True)
countries_table = countries_table.replace(to_replace = 'Russische Föderation', value = 'Russland', regex = True)

countries_table = countries_table.sort_values(by='countryCode')
countries_table.reset_index(inplace=True, drop=True)
countries_table.to_json(r'countries.json', orient='index', force_ascii=False)

countries_table
```
